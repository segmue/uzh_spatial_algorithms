## Solution 3: Line Segments and Intersections

I've added a method to calculate the determinant and then work out to which side a point is from two other points in the Point class.

I've also added a simple Bounding Box class - note that intersection is not the same as touching, and I have tried to be consistent - not trivial.

The Segment class treats equality as being independent of direction. It has an intersection method. I test it with lots of cases.

Remember, the solution is one possible approach, and I make some assumptions. Yours may be different!!

```python
# SETUP
# =====
import matplotlib.pyplot as plt
from geospatial import Point, Segment, PointGroup, Polygon, Bbox

# SAMPLE DATA
# ===========
# each item in list represents lines of 2 points with x,y coords. 
# Assume order is p.x, p.y, q.x, q.y and assume Cartesian coordinates
data = [(3,3,6,6),
        (3,6,6,3), 
        (3,1,7,7),
        (3,3,6,3), 
        (3,1,7,1),
        (3,3,6,6),
        (5,5,8,8),
        (1,10,10,10),
        (5,10,11,10),
        (4,4,10,10),
        (9,9,11,11),
        (4,4,4,8),
        (4,3,4,1)
       ]
print(f"number of lines in data: {len(data)}")
```

```python
# function to plot segment properties
def plotSegments(seg1, seg2):
    
    print(seg1)
    print(seg2)
    
    print(f"\nSegment 1 length is: {seg1.length:.2f}")
    print(f"Segment 2 length is: {seg2.length:.2f}")
    print(f"Intersection: {seg1.intersects(seg2)}")
    # plot the two lines for visual confirmation
    x1 = [seg1.start.x, seg1.end.x]
    y1 = [seg1.start.y, seg1.end.y]
    plt.plot(x1, y1, linestyle="dashed", marker="o")
    x2 = [seg2.start.x, seg2.end.x]
    y2 = [seg2.start.y, seg2.end.y]
    plt.plot(x2, y2, linestyle="dotted", marker="D")
    plt.show()
    print("---------------------------------------------------------\n\n")
```

```python
# Loop through all the segments, ignoring those which are obviously identical.

for a in range(0, len(data)):
    for b in range(0, len(data)):
        if (a != b):
                # Create segments
                seg1 = Segment(Point(data[a][0], data[a][1]), Point(data[a][2], data[a][3]))
                seg2 = Segment(Point(data[b][0], data[b][1]), Point(data[b][2], data[b][3]))
                #Plot segments
                plotSegments(seg1, seg2)
```

# 09: Visualising data

Using built-in capabilities and additional packages Matplotlib and Seaborn

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# simple line plot
xs = [1,2,3,4,5]
ys = [5,10,15,20,25]

plt.plot(xs, ys)      # must be same number of elements in each list
plt.show()
```

```python
# plot without a line
plt.plot(xs,ys,'o')
plt.show()
```

```python
# specify line style and marker style (and plotting both)
plt.plot(xs,ys, linestyle='--', linewidth=2, color='purple', marker='s', ms=5, mfc='black')
plt.show()
```

```python
# scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,84,86])

plt.scatter(x,y)
plt.show()
```

```python
# scatter plot with defined sizes of scatter points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,84,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y, s=sizes)
plt.show()
```

```python
# bar chart
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y,width=1.2)
plt.show()
```

```python
# horizontal bar chart
plt.barh(x,y,height=1.2)
plt.show()
```

```python
# histogram - generating a random normal distribution, mean = 170, sd = 10, 250 values
x = np.random.normal(170,10,250)

plt.hist(x, bins=5)
plt.show()
```

```python
# add title, figuresize, axis label, xlim, ylim and save the figure

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,84,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

font1 = {'family':'serif', 'color':'blue', 'size':20}

plt.figure(figsize=(10,10))
plt.scatter(x,y, s=sizes)
plt.title('Example', fontdict=font1, loc='left')
plt.xlabel('X Value', fontdict=font1)
plt.ylabel('Y Value', fontdict=font1)
plt.grid(axis='y')
plt.xlim(0,20)
plt.ylim(60,120)
plt.show()

plt.savefig('my_image.png')
```

# 04: Classes and methods

```python
# Define a class
class Person:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
        
p1 = Person('John', 39)

print(p1.name)
print(p1.age)

p1.age = 40
print(p1.age)
```

```python
# include a method
class Person:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def printName(self):
        print('Hello my name is '+self.name)
        
        
p1 = Person('John', 39)
print(p1.name)

p1.printName()
```

```python
# define a child class
class Student(Person):
    
    def __init__(self,name,age,gyear):
        super().__init__(name,age)
        self.graduationyear = gyear
        
        
ps1 = Student('Alana', 24, 2019)

print(ps1.graduationyear)
```

```python
# representation
class Person:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __repr__(self):
        return f'Person(name = {self.name}, age = {self.age})'

    
p1 = Person('John', 39)

p1
```

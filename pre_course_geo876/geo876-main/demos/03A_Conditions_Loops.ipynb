## Loops

```python
# while loop
i = 1
while i <=5:
    print(f'Number {i}')
    i += 1
```

```python
# for loop
my_list = ['fred', 'harry', 'sue', 'mary']

for x in my_list:           # x will be set to each value in the list    
    print(x)
```

```python
for x in range(0,len(my_list)):   # x will be the index position   
    print(my_list[x])
```

```python
for i in range(10):   # default start at 0, increment of 1
    print(i)
```

```python
for i in range(0,100,5):    # range 0 to 100, increments of 5
    print(i)
```

```python
# Nested loop
pre_fixes = ['hello', 'goodbye']
words = ['world', 'everybody']

for p in pre_fixes:
    for s in words:
        print(p, s)
```

```python
# Controlling exits from loops
my_list = ['yellow', 'blue', 'red', 'green', 'green', 'black']
for i in range(0, len(my_list)):
    item = my_list[i]
    if item == 'green':
        my_list[i] = item + '_grass'
        continue
    if item == 'blue':
        my_list[i] = item + '_sky'
        continue
    my_list[i] = item + '_other'
print(my_list)
```

```python
i = 1
while i < 10:
    print(i)
    if i == 3:
        break
    i += 1
```

# 03B: Introduction to functions

```python
# Defining a function
def myFunction():
    print('Hello')
    
myFunction()
```

```python
# Passing a collection of data into a function
def showColors(color_list):
    print(f'number of colors: {len(color_list)}')
    for c in color_list:
        print(c)

my_colors = ['red', 'green', 'blue']

showColors(my_colors)
```

```python
# Additional example not shown in tutorial
def showColors(*args):
    print(f'number of colours: {len(args)}')
    for c in args:
        print(c)

my_colors = ['red', 'green', 'blue', 'yellow']
showColors(my_colors)
```

```python
# Named function vs. Lambda
def multiAB(a, b):
    return a * b

print(multiAB(5, 6))


x = lambda a,b : a * b

print(x(5,6))
```

```python
# transforming or formatting data
full_name = lambda first, last : f'Full name is {first.title()} {last.title()}'

print(full_name('obi', 'kenobi'))
```

```python
# namespaces
def calcBill(bill):
    x = 0.02
    return bill + (bill *x)

x = 50
calcBill(x)
```

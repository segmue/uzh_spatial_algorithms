Single variables; multi-value variables (arrays); data types; printing with formatting or replacement; copying; indexing and referencing

## Multi-value variables

```python
# Lists
my_list = ['red', 'green', 'blue']
print(my_list)

my_list.append('pink')   # append value to list
print(my_list)

my_list.insert(1, 'cyan')   # insert value at specific position
print(my_list)

a_list = ['yellow', 'blue']
my_list.extend(a_list)   # extend list with other list
print(my_list)

this_list = my_list + a_list  # add lists together
print(this_list)

print(my_list[0]) 

print(my_list[2:4])  

print(my_list[:4])

my_list[2] = 'purple'
print(my_list)

my_list[3:5] = ['black', 'white']
print(my_list)

my_list.remove('blue')
print(my_list)

my_list.pop(1)
print(my_list)

del my_list[0]
print(my_list)

my_list.sort()
print(my_list)

my_list.sort(reverse = True)
print(my_list)

my_list.reverse()
print(my_list)

another_list = my_list.copy()

another_list.clear()
print(another_list)
```

```python
# dictionaries
my_dict = {'x':'Harry', 'y':'2007', 'z':['white', 'black']}
print(my_dict)

my_dict['x'] = 'Fred'
print(my_dict)

print(my_dict['x'])

my_dict.keys()

my_dict.update({'y':'2011', 'z':['red', 'blue']})
print(my_dict)

my_dict.update({'nr':45})
print(my_dict)

x = my_dict.get("y")
print(x)

my_dict.pop('x')
print(my_dict)

my_dict.clear()
print(my_dict)

del my_dict
```

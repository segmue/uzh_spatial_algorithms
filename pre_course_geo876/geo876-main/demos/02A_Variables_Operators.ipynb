# 02A: Variables and data types
Single variables; multi-value variables (arrays); data types; printing with formatting or replacement; copying; indexing and referencing

## Single-value variables

```python
# numbers and strings
x = 5
print(x, type(x))

x = 'red'
print(x, type(x))
```

```python
# numbers and strings
x = '5'       # quotes means this number is treated as a string
y = int(x)    # convert string to an integer
z = float(y)  # convert integer to a float
a = str(z)    # convert float to a string

print(x, type(x))
print(y, type(y))
print(z, type(z))
print(a, type(a))
```

```python
# operators
x = 7
y = 5

print(x/y)  # divide
print(x*y)  
print(x>y) 
print(x==3)
x += 3
print(x)
print(x > y and y < 4)
```

## Print

```python
print('x has the value ' + str(x) + ' and not ' + str(y))
print('x has the value %s and not %d' %(x,y))
print(f'x has the value {x} and not {y}')
```
